# 5.Cross join

```python
eg.


select * from emp_2 inner join dept on emp_2.did=dept.dept_id;  # inner join

select * from emp_2 left join dept on emp_2.did=dept.dept_id;   # left join

select * from emp_2 right join dept on emp_2.did=dept.dept_id;  # right join

select * from emp_2 left join dept on emp_2.did=dept.dept_id    # full outer join
union
select * from emp_2 right join dept on emp_2.did=dept.dept_id;

select * from emp_2 cross join dept                             # Cross join
```

# Constraints

```python
# syntax:
CREATE TABLE table_name (
    column1 datatype constraint,
    column2 datatype constraint,
    column3 datatype constraint,
    ....
);
```

# 1)	Not null 

```python
# example while creating table
Create table emp_not_null(empid int not null, empname varchar(10));

# example when table is already created
alter table emp modify empid int not null;
```

# 2.Unique

```python
By default, a column can have duplicate values.
Duplicates are not allowed,but null values are allowed.
```

```python
# example while creating table
Create table emp_uniq(empid int unique, empname varchar(10));
Create table emp_uniq(empid int, empname varchar(10),unique(empid));

# example when table is already created
alter table emp add constraint my_uniq unique(empid);
```

# 3.Check

```python
you can add a constraint on column values.
```

```python
# example while creating table
Create table emp_chk(empid int, empname varchar(10),age int check (age>20));
Create table emp_chk(empid int, empname varchar(10),age int check (age between 20 and 30));

# example when table is already created
alter table emp_chk add constraint my_chk check(age>20);
```

# 4.Primary key

```python
# example while creating table
Create table emp_pk(empid int primary key, empname varchar(10));
Create table emp_pk(empid int, empname varchar(10),primary key(empid));

# example when table is already created
alter table emp add constraint my_pk primary key(empid);
```

# 5.Default

```python
# example while creating table
create table emp_def(empid int default 1005,empname varchar(10));

# example when table is already created
alter table emp_df alter empid set default 1001;
```

# 6.Drop constraint

```python
Delete exisiting constraint from a table.

# example
alter table emp drop constraint my_chk;
```

# 7.Auto increment

```python
# example while creating table
create table emp_AI(empid int auto_increment primary key,empname varchar(20));

# example when table is already created
alter table emp_AI auto_increment=1001;
```

# 8.Foreign key

```python
It is used to define relationship between 2 tables
To maintain data integrity we use concept of foreign key
```

```python
# Student table and course table

create table student (roll_no int , name varchar(10),primary key (roll_no));  # parent/referenced table

create table course (course_id int,crs_name varchar(15),roll_no int,primary key (course_id),
foreign key (roll_no) references student(roll_no));    # child table/referencing table
```
